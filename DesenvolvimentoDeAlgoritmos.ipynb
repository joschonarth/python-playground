> ### `1.` Criar um programa para somar os números inteiros entre **i** e **f** (incluindo i e f), com **i** e **f** fornecidos pelo usuário.
+ ### Contador e Acumulador
+ ### Através do while

```python
# usando o while

i = int(input('Início: '))
f = int(input('\nFim: '))

ci, cf = i, f   # cópia para ser mostrado no final

if i > f:   # trocar i com f se i for maior
    i, f = f, i

c = i  # c é o contador e i é o início do intervalo
s = 0  # s é o acumulador (0 é o elemento neutro da soma)
while c <= f:  # f é o fim do intervalo
    s += c  # s = s + c (s é o acumulador)
    c += 1  # incrementa o contador

print(f'\nA soma entre {ci} e {cf} é igual a {s}')
```

> ### `Outra forma`: usando o for.

```python
# usando o for

i = int(input('Início: '))
f = int(input('\nFim: '))

s = 0  # soma (acumulador)
for c in range(i, f + 1):
    s += c  # s = s + c

print(f'\nA soma entre {i} e {f} é igual a {s}')
```

> ### **`14.`** Crie um programa para ler um número inteiro e, após a leitura, mostrar se o número é PRIMO ou não.
### `Regra matemática`: número primo é divisível por 1 e por ele mesmo, APENAS. ** Exceção: número 1.
### `Regra na programação`: todos são primos e devemos verificar se há divisores entre 2 e a metade inteira (quociente da divisão por 2) do número escolhido pelo usuário. Se houver, NÃO é primo.

```python
n = int(input('Número inteiro: '))
primo = True  # flag para afirmar que n é primo
for x in range(2, n // 2 + 1):  # percorre os possíveis divisores de n
    if n % x == 0:
        primo = False
        break  # sair da repetição, pois já encontrou um divisor

if primo and n != 1:
    print(f'\n{n} é primo')
else:
    print(f'\n{n} não é primo')
```

> ## **Jogo**: acerte o número.
---
### Regras:
### `CPU`: gerar um número aleatório entre 1 e 100.
### `Jogador`: o objetivo é acertar o número gerado no menor número de tentativas possível.
### Detalhes:
+ #### A cada tentativa com erro, o jogo deve informar se o valor "pensado" pela CPU é **menor** ou **maior** que o valor digitado pelo jogador.
+ #### Após o acerto, mostrar a qtde de tentativas que o jogador realizou e, durante o jogo, mostrar qual a tentativa o jogador está realizando no momento.

> ### `Extra`: criar um programa com uma função para calcular e retornar as raízes de uma equação de 2º grau, recebendo os termos a, b e c da equação. Se o delta for negativo, não calcular e retornar -1 nas raízes. O programa principal deve ler os termos a, b e c, chamar a função e mostrar os resultados retornados.
> ### delta = b² - 4ac
> ## $x1 = -b + \sqrt{delta}\over {2a}$
> ## $x2 = -b - \sqrt{delta}\over {2a}$

```python
def eq2g(pa, pb, pc):
    delta = pb ** 2 - 4 * pa * pc
    if delta >= 0:
        x1 = (-pb + delta ** 0.5) / (2 * pa)
        x2 = (-pb - delta ** 0.5) / (2 * pa)
    else:
        x1 = x2 = -1
    return x1, x2

def main():
    a = float(input('a: '))
    b = float(input('b: '))
    c = float(input('c: '))
    x1, x2 = eq2g(a, b, c)
    print(f'\nx1 = {x1}\nx2 = {x2}')

main()
```

media = média aritmética entre o menor e o maior valor dos 3 parâmetros (pa, pb e pc).

s = raiz(pa) . media + pb . pc

r = (s – s) / (s + s)

1) Criar um programa que gera duas listas com valores inteiros e aleatórios entre 1 e 10. A primeira lista deve ter 10 valores e a segunda 20 valores. Após as gerações, verificar e mostrar se todos os valores da 1ª lista estão contidos na 2ª lista. Mostrar apenas uma mensagem de Sim ou Não.

2) Criar um programa para gerar uma lista (vetor) com 100 valores aleatórios inteiros entre 10 e 20 no programa principal e, após a geração, chamar uma função que recebe a lista como parâmetro, calcula e retorna a quantidade de valores múltiplos de 10 contidos na lista e a média aritmética dos 50 últimos valores da lista. Os resultados retornados pela função devem ser mostrados pelo programa principal no final do programa.

```python
# Questão 1

from random import randint

lis1 = [randint(1, 10) for _ in range(10)]
lis2 = [randint(1, 10) for _ in range(20)]
sim = True
for v in lis1:
    if v not in lis2:  # verifica se o valor da lista 1 NÃO está na lista 2
        sim = False
        break
if sim:
    print('SIM')
else:
    print('NÃO')

# Questão 2

from random import randint

def calc(plis):
    q10 = 0
    for v in plis:
        if v % 10 == 0:
            q10 += 1
    med50 = sum(plis[-50:]) / 50  #  ou [50:]
    return q10, med50

def main():
    lis = [randint(10, 20) for _ in range(100)]
    qtde, media = calc(lis)
    print(qtde, media)

main()
```
