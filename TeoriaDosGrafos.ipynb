```python
import numpy as np

def mostrar_grafo(mat, grau):
    nv = mat.shape[0]  # número de vértices
    print('\nMatriz de Adjacências para Grafo:')
    print('\nV |', end='')  # mostra os vértices de destino
    for v in range(nv):
        print(f' {v + 1} ', end='')
    print(' Grau')
    print('-' * (nv * 3 + 9))
    for v in range(nv):  # percorre os vértices
        print(f'{v + 1} |', end='')  # mostra os vértices de origem
        for i in range(nv):
            print(f' {mat[v][i]} ', end='')  # mostra a matriz (0's e 1's)
        print(f'  {grau[v]}')

def mostrar_digrafo(mat, ge, gr):
    nv = mat.shape[0]  # número de vértices
    print('\nMatriz de Adjacências para Dígrafo:')
    print('\nV |', end='')  # mostra os vértices de destino
    for v in range(nv):
        print(f' {v + 1} ', end='')
    print(' GE  GR')
    print('-' * (nv * 3 + 11))
    for v in range(nv):  # percorre os vértices
        print(f'{v + 1} |', end='')  # mostra os vértices de origem
        for i in range(nv):
            print(f' {mat[v][i]} ', end='')  # mostra a matriz (0's e 1's)
        print(f'  {ge[v]}   {gr[v]}')

grafo = input('\nArestas do grafo (origem destino, origem destino, ...): ')
grafo = np.array([[int(v) for v in ar.split()] for ar in grafo.split(',')])  # cria a lista de arestas
nv = grafo.max() # calcula o maior vértice (nv: número de vértices)
matgr = np.zeros((nv, nv), dtype=np.int8)  # matriz para grafo
matdi = np.zeros((nv, nv), dtype=np.int8)  # matriz para dígrafo
grau = [0] * nv  # lista com os graus de cada vértice (grafos)
gem = [0] * nv  # lista com os graus de emissão (dígrafos)
gre = [0] * nv  # lista com os graus de recepção (dígrafos)
for ar in grafo:  # percorre a lista de arestas
    o, d = ar[0], ar[1] # aresta divida em origem e destino
    matgr[o - 1, d - 1] = matgr[d - 1, o - 1] = 1  # inserção espelhada em grafos
    matdi[o - 1, d - 1] = 1  # inserção em único sentido em dígrafos
    grau[o - 1] += 1  # aumenta o grau do vértice de origem
    grau[d - 1] += 1  # aumenta o grau do vértice de destino
    gem[o - 1] += 1  # aumenta o grau de emissão
    gre[d - 1] += 1  # aumenta o grau de recepção
mostrar_grafo(matgr, grau)
mostrar_digrafo(matdi, gem, gre)
```

```python
def busca_prof(lis, vi):
    lis[0][vi - 1] = 1  # marcar o vértice
    print(f'Marcar vértice {vi}')
    for v in lis[vi]: # percorre a lista do vértice vi (v é o vértice atual da lista)
        if lis[0][v - 1] == 0: # verifica se já foi marcado (0 - não e 1 - sim)
            busca_prof(lis, v) # busca novamente pelo vértice v (recursão)
        else:
            print(f'Vértice {v} já marcado')

vini = int(input('Vértice inicial: '))
print(f'\nBusca em profundidade com início pelo vértice {vini}:\n')
liscp = lisadj.copy()  # criar uma cópia da lista que contém as listas de adjacências
liscp.insert(0, [0] * len(liscp))  # insere a lista de marcações (0 é não marcado)
busca_prof(liscp, vini) # lis_adj foi criada na célula anterior
```

```python
def busca_larg(lis, vi):
    fila = [vi] # cria a fila de próximas buscas com o vértice inicial
    lis[0][vi - 1] = 1  # marca o vértice inicial da busca
    print(f'Marcar vértice {vi}.\t\tFila: {fila}', end='')
    while fila: # enquanto tiver valores na fila de próximas buscas
        vert = fila.pop(0) # retira da fila o próximo vértice da busca (vert)
        print(f'  -  Retirar {vert}', end='')
        for v in lis[vert]: # percorre a lista do vértice vert (v é o vértice atual da lista)
            print()
            if lis[0][v - 1] == 0: # verifica se já foi marcado (0 - não e 1 - sim)
                lis[0][v - 1] = 1  # marcar o vértice v da busca
                fila.append(v) # insere o vértice na fila de próximas buscas
                print(f'Marcar vértice {v}.\t\tFila: {fila}', end='')
            else:
                print(f'Vértice {v} já marcado.\t\tFila: {fila}', end='')


vini = int(input('Vértice inicial: '))
print(f'\nBusca em largura com início pelo vértice {vini}:\n')
liscp = lisadj.copy()  # criar uma cópia da lista que contém as listas de adjacências
liscp.insert(0, [0] * len(liscp))  # insere a lista de marcações (0 é não marcado)
busca_larg(liscp, vini) # lis_adj foi criada na célula anterior
```

>### Listas de adjacências com os dados do grafo (grafo valorado):

```python
import numpy as np

grafo = input('Arestas do grafo (origem destino custo, origem destino custo, ...): ')
grafo = np.array([[int(v) for v in ar.split()] for ar in grafo.split(',')])
nv = grafo[:, :2].max() # calcula o maior vértice (número de vértices)
lis_adj = [[] for _ in range(nv)]
for ar in grafo:
    ori, des, custo = ar[0], ar[1], ar[2] # dados de uma aresta
    lis_adj[ori - 1].append([des, custo]) # insere chegada na saída
    if ori != des: # se não for um laço (vs igual a vc)
        lis_adj[des - 1].append([ori, custo]) # insere saída na chegada
print('\nListas de adjacências com custos em cada aresta:\n')
for i, lis in enumerate(lis_adj):
    print(f'{i + 1} : {lis}')
```

> ### `Algoritmo de Dijkstra`: construção da tabela de dados (custos mínimos e vértices anteriores).

```python
def dijkstra(lis, vi):
    dados = [[float('inf'), ''] for i in range(len(lis))] # tabela de dados
    dados[vi - 1][0], dados[vi - 1][1] = 0, '-' # custo zero para o vértice inicial
    abertos = [i + 1 for i in range(len(lis))] # lista de vértices abertos
    while abertos: # enquanto tiver vértices abertos
        # calcular vértice com menor custo
        mc = float('inf')
        for v in abertos:  # busca pelo vértice com menor custo dos vértices em abertos
            if dados[v - 1][0] < mc:
                mc = dados[v - 1][0]
                vef = v  # armazena o vértice em fechamento (com menor custo)
        abertos.remove(vef) # remove o vértice com menor custo dos vértices abertos
        cbase = mc # custo do vértice em fechamento (custo base)
        for ar in lis[vef - 1]: # percorre a lista do vértice em fechamento (vértice e custo)
            v, c = ar[0], ar[1] # vértice e custo de cada aresta da lista em fechamento
            if v in abertos and cbase + c < dados[v - 1][0]: # verifica se o custo até o vértice atual é menor
                dados[v - 1][0] = cbase + c # atualiza o custo mínimo até v
                dados[v - 1][1] = vef # atualiza o vértice anterior (vef - vértice em fechamento)
    return dados

vi = int(input('\nVértice inicial (Dijkstra): '))
dados = dijkstra(lis_adj, vi)
print('\nTabela de dados Dijkstra:\n')
for i, lin in enumerate(dados):
    print(f'{i + 1:2}: {lin}')
print('\nFim...')
```
